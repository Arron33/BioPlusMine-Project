```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits import mplot3d
```

```python
df = pd.read_csv('PhilippinesData.csv')
```

```python
df.hist(bins=50, figsize=(20,15))
```

```python
fig = plt.figure(figsize=(10,8))
sns.heatmap(data=df.corr());
plt.plot;
```

```python
colour_list = []

for i in range(132):
    if df['Q'][i] == 'Q1':
        colour_list.append('r')
    elif df['Q'][i] == 'Q2':
        colour_list.append('g')
    elif df['Q'][i] == 'Q3':
        colour_list.append('b')
    elif df['Q'][i] == 'Q4':
        colour_list.append('orange')
    else:
        colour_list.append('gray')

print(colour_list)
```

```python
title = 'Waste NDVI v Ferrous Min'
x = 'Waste Ferrous Minerals'
y = 'Waste NDVI'

fig = plt.figure(figsize = (8,6),dpi=150)
ax = fig.add_subplot(1, 1, 1)
ax.set_title(title)
ax.set_xlabel(x)
ax.set_ylabel(y)


for i in range(132):
    plt.scatter(df[x][i], df[y][i], c = colour_list[i])

plt.plot(df[x], df[y], c='gray', linewidth=1, zorder=-1)


plt.legend(bbox_to_anchor=(1.04,0, 0.45,1), loc="upper left", mode="expand", borderaxespad=0, ncol=1)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='grey')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='grey')
plt.plot;
plt.show()
```
